## Part 2: UMAP Visualization (Static)

```python
# color dictionary (same color palette used in Glinos paper)
color_dict = {"Adipose - Subcutaneous":"#ffc0cb",
             "Brain - Anterior cingulate cortex (BA24)":"#8d5b96",
             "Brain - Caudate (basal ganglia)":"#7776b1",
             "Brain - Cerebellar Hemisphere":"#9773ba",
             "Brain - Frontal Cortex (BA9)":"#b873ba",
             "Brain - Putamen (basal ganglia)":"#c893c9",
             "Breast - Mammary Tissue":"#ff69b4",
             "Cells - Cultured fibroblasts":"#d4a910",
             "Heart - Atrial Appendage":"#c4625d",
             "Heart - Left Ventricle":"#bc3c28",
             "Liver":"#815375",
             "Lung":"#0072b5",
             "Muscle - Skeletal":"#1f854e",
             "Pancreas":"#e18726"}
```

```python
def visualize_for_UMAP_static(df,num_neighbors,min_distance,figure_title):
    df_for_decomp = df.iloc[:,2:] #transcript quantification data only
    df_for_decomp = df_for_decomp.astype('float') #change data type from object to numeric
    # scale the transcript quantification data before UMAP
    scaler = StandardScaler()
    df_for_decomp_sc = scaler.fit_transform(df_for_decomp)
    reducer = umap.UMAP(random_state=107,n_neighbors=num_neighbors,min_dist=min_distance)
    embedding = reducer.fit_transform(df_for_decomp_sc)
    # organize UMAP embeddings into a dataframe and add tissue type information
    transformed_df = pd.DataFrame(embedding,columns=['Embed1', 'Embed2'])
    transformed_df['tissue'] = annotated_transcript_express_tissue_clean['tissue']
    # plot UMAP in 2 dimensions
    plt.figure(figsize=(15, 10))
    for key, group in transformed_df.groupby(['tissue']):
        plt.plot(group.Embed1, group.Embed2, 'o', markersize=10, 
             c=color_dict.get(key), label=key)
    plt.legend(bbox_to_anchor=(1.02, 0.5), loc="center left",fontsize=12)
    plt.xlabel('Embedding 1',fontsize=16)
    plt.ylabel('Embedding 2',fontsize=16)
    plt.title(figure_title,fontsize=18)
    plt.grid()
```

### Novel Transcripts (No Filter)

```python
visualize_for_UMAP_static(novel_transcript_express_tissue_clean,25,0.1,
                         str(num_novel_transcripts)+' Novel Transcripts')
```

### Novel Transcripts (Filter Applied)

```python
visualize_for_UMAP_static(novel_transcript_express_tissue_clean_filtered,25,0.1,
                         str(num_novel_filtered_transcripts)+' Novel Transcripts')
```

### Annotated Transcripts (No Filter)

```python
visualize_for_UMAP_static(annotated_transcript_express_tissue_clean,25,0.1,
                         str(num_annotated_transcripts)+' Annotated Transcripts')
```

### Annotated Transcripts (Filter Applied)

```python
visualize_for_UMAP_static(annotated_transcript_express_tissue_clean_filtered,25,0.1,
                         str(num_annotated_filtered_transcripts)+' Annotated Transcripts')
```

## Part 3: UMAP Visualization (Interactive)

```python
from bokeh.models import Title
```

```python
from bokeh.models import Legend
```

```python
def visualize_for_UMAP_interactive(df,num_neighbors,min_distance,figure_title):
    df_for_decomp = df.iloc[:,2:] #transcript quantification data only
    df_for_decomp = df_for_decomp.astype('float') #change data type from object to numeric
    # scale the transcript quantification data before UMAP
    scaler = StandardScaler()
    df_for_decomp_sc = scaler.fit_transform(df_for_decomp)
    mapper = umap.UMAP(random_state=107,n_neighbors=num_neighbors,min_dist=min_distance).fit(df_for_decomp_sc)
    hover_data = pd.DataFrame({'tissue':annotated_transcript_express_tissue_clean['tissue'],
                        'sample_id':annotated_transcript_express_tissue_clean['sample_id']})
    # interactive plot
    umap.plot.output_notebook()
    # map the tissue categories to the corresponding color in the color dictionary
    colors = [color_dict[tissue] for tissue in hover_data.tissue]
    p = umap.plot.interactive(mapper, hover_data=hover_data, labels=hover_data.tissue,color_key=color_dict,point_size=10)
    # add title
    p.title = Title(text=figure_title, align="center")
    p.add_layout(p.title, "above")
    umap.plot.show(p)
```

### Novel Transcripts (No Filter)

```python
visualize_for_UMAP_interactive(novel_transcript_express_tissue_clean,25,0.1,
                              str(num_novel_transcripts)+' Novel Transcripts')
```

### Novel Transcripts (Filter Applied)

```python
visualize_for_UMAP_interactive(novel_transcript_express_tissue_clean_filtered,25,0.1,
                              str(num_novel_filtered_transcripts)+' Novel Transcripts')
```

### Annotated Transcripts (No Filter)

```python
visualize_for_UMAP_interactive(annotated_transcript_express_tissue_clean,25,0.1,
                              str(num_annotated_transcripts)+' Annotated Transcripts')
```

### Annotated Transcripts (Filter Applied)

```python
visualize_for_UMAP_interactive(annotated_transcript_express_tissue_clean_filtered,25,0.1,
                              str(num_annotated_filtered_transcripts)+' Annotated Transcripts')
```
